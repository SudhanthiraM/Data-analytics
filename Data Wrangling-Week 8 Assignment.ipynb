### 1.Read in the covid19_cases.csv file

```python
import pandas as pd
url="https://raw.githubusercontent.com/svkarthik86/Advanced-python/main/WEEK-8%20Assignment/covid19_cases.csv"
covid19_cases=pd.read_csv(url)
covid19_cases
```

```python
type(covid19_cases)
```

### 2.Create a date column using the data in the dateRep column and the pd.to_datetime() function.


```python
covid19_cases['Date']=pd.to_datetime(covid19_cases[["year","month","day"]])
covid19_cases
```

### 3.Set the date column as the index and sort the index.


```python
covid19_cases_pivot=covid19_cases.pivot(index="Date",columns="countriesAndTerritories",values="cases").fillna(0)
covid19_cases_pivot
```

### 4.Replace occurrences of United_States_of_America and United_Kingdom with USA and UK, respectively.

```python
covid19_cases_pivot.columns=covid19_cases_pivot.columns.str.replace("United_States_of_America","USA")
covid19_cases_pivot.columns=covid19_cases_pivot.columns.str.replace("United_Kingdom","UK")
covid19_cases_pivot
```

### 5.Using the countriesAndTerritories column, filter the data down to Argentina, Brazil, China, Colombia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA.

```python
covid19_cases_pivot.filter(["Argentina", "Brazil", "China", "Colombia", "India", "Italy", "Mexico", "Peru", "Russia", "Spain", "Turkey", "UK", "USA"])
covid19_cases_pivot
```
